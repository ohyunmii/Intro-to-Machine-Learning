```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 
%matplotlib inline
```

```python
sns.set_style('whitegrid') #document에 스타일 관련 내용이 자세히 쓰여있음
```

```python
x = np.linspace(0, 14, 100) #0부터 14까지 100개

y1 = np.sin(x)
y2 = 2 * np.sin(x + 0.5)
y3 = 3 * np.sin(x + 1.0)
y4 = 4 * np.sin(x + 1.5)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips) # hue='sex'
plt.show()
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights = flights.pivot("month", "year", "passengers") #index로 만들 column명, column으로 만들, 채워줄 데이터
flights.head()
```

```python
# Heat map
# 연속적인 데이터를 시각화할 때 유용함. 서로간의 연관관계가 깊을 때.
# 데이터의 강도, 농도를 보고싶을 때 사용.
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt='d') # annot: 맵에 숫자 표시, fmt: 표현 형식 
                                            # d: 정수형태, 기본은 실수형태(f)로 표현
plt.show()

# 응용: 머신러닝에서 hyperparameterd에 따른 정확도를 볼 때 사용할 수 있겠지?
```
